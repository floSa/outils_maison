```python
import re
import shutil
from pathlib import Path
from typing import List

def natural_key(text: str) -> List:
    """Convertit une chaîne en une liste alternant chaînes et entiers pour un tri naturel."""
    return [int(part) if part.isdigit() else part.lower() for part in re.split(r'(\d+)', text)]

def reorder_pages(
    input_dir: str,
    output_dir: str,
    file_name: str,
    ascending: bool,
    pas: int,
    page_num: int
) -> List[Path]:
    """
    Réordonne des images selon l'ordre naturel (comme l'explorateur Windows),
    puis attribue des numéros de page.

    Paramètres
    ----------
    input_dir : dossier contenant les images (jpg/png)
    output_dir : dossier de sortie
    file_name : préfixe pour les noms de sortie (ex: "dicobat_")
    ascending : True = numéros croissants (page_num, page_num+pas, ...)
                False = numéros décroissants (page_num, page_num-pas, ...)
    pas : écart entre les numéros
    page_num : numéro de la première page (celle du premier fichier dans l'ordre naturel)
    """
    out_path = Path(output_dir)
    out_path.mkdir(parents=True, exist_ok=True)

    extensions = {'.jpg', '.jpeg', '.png'}
    files = [f for f in Path(input_dir).iterdir()
             if f.is_file() and f.suffix.lower() in extensions]

    if not files:
        raise ValueError(f"Aucune image trouvée dans {input_dir}")

    # Tri selon l'ordre naturel (comme l'explorateur Windows)
    files.sort(key=lambda f: natural_key(f.name))

    # Afficher l'ordre pour vérification
    print("Ordre naturel des fichiers (identique à l'explorateur Windows) :")
    for i, f in enumerate(files):
        print(f"  {i+1}: {f.name}")

    step = pas if ascending else -pas
    print(f"Attribution des numéros : premier = {page_num}, pas = {step}")

    created = []
    for i, src in enumerate(files):
        new_num = page_num + i * step
        ext = src.suffix.lower()
        new_name = f"{file_name}{new_num}{ext}"
        dest = out_path / new_name
        shutil.copy2(src, dest)
        print(f"  {src.name} -> {new_name}")
        created.append(dest)

    return created
```
