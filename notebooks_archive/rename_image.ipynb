# Rename Folder simple

```python
import os
import re
import unicodedata

def nettoyer_correctement(dossier_path):
    extensions_images = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp')

    if not os.path.exists(dossier_path):
        print(f"Erreur : Le dossier '{dossier_path}' n'existe pas.")
        return

    count = 0
    for filename in os.listdir(dossier_path):
        if filename.lower().endswith(extensions_images):
            old_path = os.path.join(dossier_path, filename)
            
            nom, ext = os.path.splitext(filename)

            # 1. Enlever espaces debut/fin
            nom = nom.strip()
            
            # 2. Remplacer les espaces du milieu par des underscores (_)
            # pour éviter le problème "nomillisible.jpg"
            nom = nom.replace(" ", "_")

            # 3. Enlever les accents (café -> cafe)
            nom = unicodedata.normalize('NFKD', nom).encode('ascii', 'ignore').decode('utf-8')

            # 4. Enlever les caractères spéciaux (garde lettres, chiffres, tirets et underscores)
            nom = re.sub(r'[^a-zA-Z0-9_\-]', '', nom)

            # 5. TOUT METTRE EN MINUSCULE (ta demande)
            nom = nom.lower()

            new_filename = nom + ext.lower() # Extension en minuscule aussi
            new_path = os.path.join(dossier_path, new_filename)

            if old_path != new_path:
                try:
                    os.rename(old_path, new_path)
                    print(f"[OK] {filename} -> {new_filename}")
                    count += 1
                except OSError as e:
                    print(f"[Erreur] {filename} : {e}")

    print(f"Terminé : {count} fichiers renommés proprement.")


nettoyer_correctement(path)
```
