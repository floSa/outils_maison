```python
import os
import glob
from moviepy.editor import VideoFileClip, concatenate_videoclips
from pathlib import Path

def fusionner_videos_mp4(dossier_path, nom_sortie="compile.mp4"):
    """
    Fusionne toutes les vidéos MP4 d'un dossier en ordre alphabétique
    
    Args:
        dossier_path (str): Chemin vers le dossier contenant les vidéos
        nom_sortie (str): Nom du fichier de sortie (par défaut: "compile.mp4")
    """
    
    # Vérifier si le dossier existe
    if not os.path.exists(dossier_path):
        print(f"Erreur: Le dossier '{dossier_path}' n'existe pas.")
        return
    
    # Récupérer tous les fichiers MP4 du dossier
    pattern = os.path.join(dossier_path, "*.mp4")
    fichiers_mp4 = glob.glob(pattern)
    
    # Trier par ordre alphabétique
    fichiers_mp4.sort()
    
    if not fichiers_mp4:
        print("Aucune vidéo MP4 trouvée dans le dossier.")
        return
    
    print(f"Vidéos trouvées ({len(fichiers_mp4)}):")
    for i, fichier in enumerate(fichiers_mp4, 1):
        print(f"{i}. {os.path.basename(fichier)}")
    
    try:
        # Charger toutes les vidéos
        print("\nChargement des vidéos...")
        clips = []
        
        for fichier in fichiers_mp4:
            print(f"Chargement: {os.path.basename(fichier)}")
            clip = VideoFileClip(fichier)
            clips.append(clip)
        
        # Fusionner les vidéos
        print("\nFusion en cours...")
        video_finale = concatenate_videoclips(clips, method="compose")
        
        # Créer le chemin de sortie
        chemin_sortie = os.path.join(dossier_path, nom_sortie)
        
        # Exporter la vidéo fusionnée
        print(f"Export vers: {chemin_sortie}")
        video_finale.write_videofile(
            chemin_sortie,
            codec='libx264',
            audio_codec='aac',
            temp_audiofile='temp-audio.m4a',
            remove_temp=True
        )
        
        # Libérer la mémoire
        video_finale.close()
        for clip in clips:
            clip.close()
        
        print(f"\n✅ Fusion terminée ! Fichier créé: {nom_sortie}")
        
    except Exception as e:
        print(f"❌ Erreur lors de la fusion: {str(e)}")
```

```python
# Utilisation
# Remplacez par le chemin vers votre dossier
dossier_videos = r"C:/Users/FLORIAN/dwhelper/aylin_corazzina"  # Windows
# dossier_videos = "/votre/chemin/vers/le/dossier"   # Linux/Mac

# Lancer la fusion
fusionner_videos_mp4(dossier_videos, "aylin_corazzina_compile.mp4")
```
