```python
import os
import re
from mutagen.easyid3 import EasyID3

# Chemin absolu du répertoire cible
directory = r"C:\Users\FLORIAN\dwhelper\Baptiste Thiry\Love Marks"

# Contrôle du répertoire
if not os.path.exists(directory):
    print(f"Erreur : Le dossier {directory} est inaccessible.")
else:
    # Parcours des fichiers du répertoire
    for filename in os.listdir(directory):
        # Capture des deux premiers chiffres du nom du fichier
        match = re.match(r"^(\d{2}) - .*\.mp3$", filename)
        
        if match:
            num_str = match.group(1)
            current_path = os.path.join(directory, filename)
            
            # Reconstruction du nom de fichier original
            original_filename = f"Album Love Marks - Cezame Music Agency-{num_str}.mp3"
            original_path = os.path.join(directory, original_filename)
            
            # Suppression intégrale des métadonnées ID3 pour un retour à zéro
            try:
                tags = EasyID3(current_path)
                tags.delete()
                tags.save()
            except Exception:
                # Ignorer si le fichier n'a pas ou plus de tags
                pass
            
            # Application du renommage inversé
            os.rename(current_path, original_path)
            print(f"Restauration : {filename} -> {original_filename}")
```
