# trie des cotes de la bibliothèque

```python
import re
from pathlib import Path

def parse_lignes(texte):
    pattern = re.compile(r"""
        ^\s*
        (?P<artiste>.*?)        
        \s*-\s*
        (?P<album>.*?)          
        \s*-\s*
        (?P<cote>[^-]+?)        
        \s*(?:-|$)              
    """, re.VERBOSE | re.IGNORECASE)

    resultats = []
    
    for ligne in texte.strip().split("\n"):
        ligne = ligne.strip()
        if not ligne:
            continue
        
        match = pattern.match(ligne)
        if match:
            artiste = match.group("artiste").strip()
            album = match.group("album").strip()
            cote = match.group("cote").strip()
            
            resultats.append((artiste, album, cote))
    
    return resultats


def trier_par_cote(liste):
    return sorted(liste, key=lambda x: x[2])


def afficher_et_sauver(liste, fichier="cotes_bibli.txt"):
    """
    - print console
    - écriture fichier avec sauts de ligne fiables
    """
    lignes = [f"{artiste} - {album} - {cote}" for artiste, album, cote in liste]
    
    # 1. Affichage console
    print("\n".join(lignes))
    
    # 2. Écriture fichier (UTF-8 + \n explicite)
    chemin = Path(fichier)
    chemin.write_text("\n".join(lignes), encoding="utf-8")
    
    print(f"\nFichier écrit : {chemin.resolve()}")



```

```python
ligne = """
Ahmed Malek - Musique originale de films (Habibi Funk 003) - 786.2 MAL 1
Ahmed Malek - Musique originale de films. 2  - 786.2 MAL 
Ahmed Malek - Musique originale de films -     786.2 MAL 2 - Retour: 18/09/2025
Johann Johannsson - The Theory of everything - 786.1 THE 1 - 
Gustavo Santaolalla - The last of us - 786.46 LAS
Cliff Martinez - The Knick - 786.1 KNI 3 
Gustavo Santaolalla - The Last of us Part II - 786.46 LAS
Alexandre Desplat - Pinocchio - 786.1 PIN 1 
Alexandre Desplat - The Danish girl -     786.11 DAN - Retour 03/12/2025
ahmed malek - Musique originale de films - 786.2 MAL 2 
ahmed malek - Musique originale de films. 2 - 786.2 MAL 2 
Anne-Sophie Versnaeyen - La belle époque - 786.1 BEL 1
Danny Elfman - Alice through the looking glass - 786 B.O par dieu jeunesse
"""

data = parse_lignes(ligne)
data_triee = trier_par_cote(data)
afficher_et_sauver(data_triee)
```
