```python
import os
import subprocess

def convert_mp4_to_flac(input_folder, output_folder):
    # Créez un dossier pour les fichiers convertis s'il n'existe pas
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Parcourir tous les fichiers dans le répertoire donné
    for filename in os.listdir(input_folder):
        if filename.endswith(".mp4"):
            input_file = os.path.join(input_folder, filename)
            # Génère un nom pour le fichier FLAC
            output_file = os.path.join(
                output_folder, os.path.splitext(filename)[0] + ".flac"
            )

            # Utilisez FFmpeg pour la conversion
            command = [
                "ffmpeg",
                "-i", input_file,  # fichier d'entrée
                "-vn",  # désactiver la vidéo (audio uniquement)
                "-acodec", "flac",  # format de sortie FLAC
                output_file,  # fichier de sortie
            ]

            # Lancez la commande FFmpeg
            print(f"Converting {input_file} to {output_file}...")
            subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            print(f"Finished converting {filename}.")

# Exemple d'utilisation
input_folder = "C:/Users/FLORIAN/Videos/4K_Video/Feed_the_Flame/"  # Remplacez par le chemin de votre dossier avec les vidéos MP4
output_folder = "C:/Users/FLORIAN/Videos/4K_Video/Feed_the_Flame/"  # Remplacez par le chemin de votre dossier pour les fichiers FLAC
convert_mp4_to_flac(input_folder, output_folder)
```
